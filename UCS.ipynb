```python
import heapq

def find_least_cost_path(graph, start, goal):
    # Priority Queue stores: (cumulative_cost, current_node, path_taken)
    priority_queue = [(0, start, [start])]
    visited = set()

    while priority_queue:
        (cost, current_node, path) = heapq.heappop(priority_queue)

        if current_node in visited:
            continue

        visited.add(current_node)

        # If we reached the goal, return the results
        if current_node == goal:
            return cost, path

        # Check neighbors
        for neighbor, weight in graph[current_node].items():
            if neighbor not in visited:
                heapq.heappush(priority_queue, (cost + weight, neighbor, path + [neighbor]))

    return float("inf"), []

# Graph Representation
graph = {
    'S': {'A': 3, 'B': 2, 'C': 7},
    'A': {'D': 3, 'E': 8, 'G': 15},
    'B': {'G': 20},
    'C': {'G': 6},
    'D': {},
    'E': {},
    'G': {}
}

cost, path = find_least_cost_path(graph, 'S', 'G')

print(f"Least Cost: {cost}")
print(f"Path: {' -> '.join(path)}")
```
